```python
import selenium 
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings ("ignore")
from selenium.webdriver.common.by import By 
import time 
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys("Data Analyst")
```

```python
location=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
title=driver.find_elements(By.XPATH,"//h2[@itemprop='name']")

for i in title:
    title_tags=i.text
    job_title.append(title_tags)
```

```python
location=driver.find_elements(By.XPATH,"//div[@class='jobCard_jobCard_lists__fdnsc']/div/div")

for i in location:
    location_tags=i.text
    job_location.append(location_tags)
```

```python
company=driver.find_elements(By.XPATH,"//div[@class='jobCard_jobCard_cName__mYnow']")

for i in company:
    company_tags=i.text
    company_name.append(company_tags)
```

```python
experience=driver.find_elements(By.XPATH,"//div[@class='jobCard_jobCard_lists__fdnsc']/div/div")

for i in experience:
    experience_tags=i.text
    experience_required.append(experience_tags)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df=pd.DataFrame({'title':job_title,'location':job_location,'company':company_name,'experience':experience_required})
df
```

```python
df.head(10)
```
